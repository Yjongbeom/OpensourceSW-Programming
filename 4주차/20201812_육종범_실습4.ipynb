# 예제 1

```python
s = 'cafè'
print(len(s))

b = s.encode('utf8')
print(b)
print(len(b))

a = b.decode('utf8')
print(a)
```

# 예제 2

```python
cafe = bytes('cafè', encoding='utf_8')
print(cafe)

print(cafe[0])

print(cafe[:1])

cafe_arr = bytearray(cafe)
print(cafe_arr)

print(cafe_arr[-1:])
```

# 예제 5

```python
import struct

fmt = '<3s3sHH'

# 임의의 simpsons.gif 파일을 넣어 결과를 확인하였습니다.
with open('./simpsons.gif', 'rb') as fp:
    img = memoryview(fp.read())

header = img[:10]
print(bytes(header))

print(struct.unpack(fmt, header))

del header
del img
```



# 예제 6

```python
for codec in ['latin_1', 'utf8', 'utf16']:
    print(codec, 'El Niño'.encode(codec), sep='\t')

for codec in ['ascii', 'latin_1', 'cp1252', 'cp437', 'utf8', 'utf-16le', 'gb2312']:
    print(codec, 'A'.encode(codec), sep='\t')
```

# 예제 7

```python
city = 'São Paulo'
print(city.encode('utf-8'))

print(city.encode('utf_16'))

print(city.encode('iso8859_1'))
```

```python
# cp437은 ã 문자 인코딩 불가능
city.encode('cp437')
```

```python
print(city.encode('cp437', errors='ignore')) # 인코딩 에러 건너뜀
print(city.encode('cp437', errors='replace')) # 물음표로 치환
print(city.encode('cp437', errors='xmlcharrefreplace')) # XML 객체로 치환
```

# 예제 8

```python
octets = b'Montr\xe9al'
print(octets.decode('cp1252'))
print(octets.decode('iso8859_7'))
print(octets.decode('koi8_r'))
```

```python
print(octets.decode('utf8'))
```

```python
# 알 수 없는 문자를 표현하기 위해 사용하는 유니코드 공식 치환 문자
print(octets.decode('utf8', errors='replace'))

print('é'.encode('utf8'))
```
